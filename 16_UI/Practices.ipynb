```python
import tkinter as tk
```

```python
root = tk.Tk()
root.title( "My program")

frame = tk.Frame(root)
frame.config(bg="blue", width=400, height=300)
frame.pack()
root.mainloop()
```

```python
root = tk.Tk()
root.title( "My program")

text = "This is my first label"
label = tk.Label(root, text = text)
label.config(fg="green", bg="lightgray", font=("cortana", 30))
label.pack()
root.mainloop()
```

```python
def es_primo(num):
    """ Determina si un número es primo """
    if num <= 1:
        return False
    for i in range (2, num):
        if num % i == 0:
            return False

    return True
    
```

```python
es_primo(17)
```

```python
def sumar_numeros_primos(n):
    """ Suma todos los números primos menores o iguales a n """
    result = 0
    for i in range(0, n):
        if es_primo(i) == True:
            result += i
            print("Numero: " + str(i))
    
    return result
```

```python
sumar_numeros_primos(10)
```

```python
#Exercise 01
class car:
    def __init__(self, brand, color, combustible, cil):
        self.brand = brand
        self.color = color
        self.combustible = combustible
        self.cil = cil

    def show_data(self):
        print("the brand of the car is {}, it is color {}, it uses {} and cil is {}".format(self.brand, self.color, self.combustible, self.cil))
```

```python
myCar = car('Opel', 'red', 'gas', '1.6')
myCar.show_data()
```

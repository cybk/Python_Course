```python
def count_words(text) :
    words = text.replace(".", "").split()
    result = {}
    for word in words:
        if word not in result:
            result[word] = 1
        else:
            result[word] += 1

    return result
```

```python
text = "Este es un ejemplo de conteo de palabras. Este es un ejercicio simple."
resultado = count_words(text)
print(f'resultado = {resultado}')
```
